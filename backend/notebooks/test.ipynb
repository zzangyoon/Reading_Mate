```python
# Cell 1: 환경 설정
import sys
from pathlib import Path

# backend 디렉토리 추가
notebook_dir = Path.cwd()
if notebook_dir.name == "notebooks":
    backend_dir = notebook_dir.parent
else:
    backend_dir = notebook_dir
    
sys.path.insert(0, str(backend_dir))

from dotenv import load_dotenv
load_dotenv()

print(f"Backend directory: {backend_dir}")
print("환경 설정 완료!")

# Cell 2: Import 테스트
from app.core.database import DatabaseManager
from app.core.vector_store import VectorStoreManager
from app.core.system import ReadingAssistantSystem

print("Import 성공!")

# Cell 3: Database 테스트
db_manager = DatabaseManager()
book_meta = db_manager.get_book_metadata(1)
print(f"책 정보: {book_meta}")

# Cell 4: Vector Store 테스트
vs_manager = VectorStoreManager(db_manager)
result = vs_manager.hybrid_search("사이클론", "뭐야?", k=3)
print(f"검색 결과: {result['book_title']}")
print(result)

# Cell 5: 전체 시스템 테스트
assistant = ReadingAssistantSystem()
answer = assistant.ask("그 집을 사이클론의 한가운데로 끌어올렸다", "여기서 사이클론이 의미하는게 뭐야?")
print(answer)
```
